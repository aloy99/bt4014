```python
def process_data_top_n(path, n, output_filename):
    df = pd.read_csv(path)
    df = pd.get_dummies(df, columns=['user_feature_0', 'user_feature_1', 'user_feature_2', 'user_feature_3'], dtype=float)
    n_most_popular_items = df.groupby(['item_id']).agg({
        'click': 'mean'
    }).sort_values('click', ascending = False).iloc[:n].index.to_list()
    mapping = dict(zip(n_most_popular_items, list(range(n))))
    df_n_filtered = df[df['item_id'].isin(n_most_popular_items)]
    df_n_filtered = df_n_filtered[df_n_filtered.columns[1:]]
    df_n_filtered['item_id'] = df_n_filtered['item_id'].map(mapping)
    df_n_filtered = df_n_filtered.drop(['timestamp', 'position'], axis=1)
    df_n_filtered.to_csv(f"top_{n}_{output_filename}.csv", index=False)
    print(n)
    print(len(df_n_filtered.columns) - 3)
```

```python
process_data_top_n(r'open_bandit_dataset\bts\men\men.csv', 15, 'bts_men')
```

```python
process_data_top_n(r'open_bandit_dataset\random\men\men.csv', 15, 'random_men')
```
