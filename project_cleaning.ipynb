```python
import pandas as pd
```

```python
df = pd.read_csv(r'bts\men\men.csv')
df = pd.get_dummies(df, columns=['user_feature_0', 'user_feature_1', 'user_feature_2', 'user_feature_3'], dtype=float)
df = df[df.columns[1:]]
df = df.drop(['timestamp', 'position'], axis=1)
df.to_csv(r"test_bts_men.csv", index=False)
df
```
